```python
from datetime import datetime
from opensky_data.impala import ImpalaWrapper

opensky = ImpalaWrapper("", "")
flight = opensky.history(datetime(2017, 11, 23), datetime(2017, 11, 24), callsign="DLH66N")
```

```python
from cartopy.feature import NaturalEarthFeature
from cartopy.crs import PlateCarree, EuroPP

countries = NaturalEarthFeature(
    category='cultural',
    name='admin_0_countries',
    scale='10m',
    edgecolor='#524c50',
    facecolor='none',
    alpha=.5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection=EuroPP())

ax.add_feature(countries)
ax.gridlines()
ax.set_extent((-7, 15, 40, 55))

ax.plot(flight.longitude, flight.latitude, transform=PlateCarree())
```
